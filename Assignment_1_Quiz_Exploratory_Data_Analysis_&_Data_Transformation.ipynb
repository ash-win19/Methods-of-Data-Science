```python
import matplotlib.pyplot as plt
import seaborn as sns
plot_parametrics = df[['Horsepower', 'MSRP']]
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Horsepower', y='MSRP', data=plot_parametrics, color= '#DB8989')
plt.title('Relationship between Horsepower and MSRP')
plt.xlabel('Horsepower')
plt.ylabel('MSRP')
plt.show()
```

```python
make_count = df['Make'].value_counts()
plt.figure(figsize=(36, 12))
sns.barplot(x=make_count.index, y=make_count.values, palette='viridis')
plt.title('Number of Vehicles by Make')
plt.xlabel('Make')
plt.ylabel('Number of Vehicles')
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df['NormalizedHorsepower'] = scaler.fit_transform(df[['Horsepower']])
print(df['NormalizedHorsepower'])
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df['StandardizedHorsepower'] = scaler.fit_transform(df[['Horsepower']])
print(df['StandardizedHorsepower'])
```
